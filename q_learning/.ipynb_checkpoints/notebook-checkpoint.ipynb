```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import scipy.stats as sts
import ast
from scipy.interpolate import interp1d
from collections import OrderedDict
from matplotlib.transforms import blended_transform_factory

plt.close('all')
plt.style.use('default')
plt.rc('font', family='serif')
plt.rc('text', usetex=True)

LEGEND_FONT_SIZE = 12
AXIS_FONT_SIZE = 16
TICKS_FONT_SIZE = 12

figs = []

def running_mean(x, N):
    divider = np.convolve(np.ones_like(x), np.ones((N,)), mode='same')
    return np.convolve(x, np.ones((N,)), mode='same') / divider
```

```python
linestyles = OrderedDict(
    [('solid',               (0, ())),
     ('loosely dotted',      (0, (1, 10))),
     ('dotted',              (0, (1, 5))),
     ('densely dotted',      (0, (1, 1))),

     ('loosely dashed',      (0, (5, 10))),
     ('dashed',              (0, (5, 5))),
     ('densely dashed',      (0, (5, 1))),

     ('loosely dashdotted',  (0, (3, 10, 1, 10))),
     ('dashdotted',          (0, (3, 5, 1, 5))),
     ('densely dashdotted',  (0, (3, 1, 1, 1))),

     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1))),

     ('loosely dashdashdotted', (0, (3, 10, 3, 10, 1, 10))),
     ('dashdashdotted',         (0, (3, 5, 3, 5, 1, 5))),
     ('densely dashdashdotted', (0, (3, 1, 3, 1, 1, 1)))])
```

```python
def plot_ci(xs, ys, conf, n, xlabel=None, ylabel=None, label=None, ax=None, **kwargs):
    #print(xs)
    all_x = np.array(list(set(np.concatenate(xs).ravel().tolist())))
    all_x = np.sort(all_x)
    last_x = min(map(lambda x: max(x), xs))
    first_x = max(map(lambda x: min(x), xs))
    all_ys = []
    
    pred_x = np.linspace(first_x, last_x, 200)
    
    for i in range(len(xs)):
        f = interp1d(xs[i], ys[i], "linear")
        all_ys.append(f(pred_x))
    all_ys = np.array(all_ys)
    
    data_mean = np.mean(all_ys, axis=0)
    data_std = np.std(all_ys, axis=0) + 1e-24
    interval = sts.t.interval(conf, n-1,loc=data_mean,scale=data_std/np.sqrt(n))

    col = kwargs.get('color')
    if ax is None:
        fig, ax = plt.subplots(figsize=(4.8, 3.2))
        figs.append(fig)
    if label is None:
        ax.plot(pred_x, data_mean, **kwargs)
    else:
        ax.plot(pred_x, data_mean, label=label, **kwargs)
        
    if col is not None:
        ax.fill_between(pred_x, interval[0], interval[1], alpha=0.2, color=col, linewidth=0.)
    else:
        ax.fill_between(pred_x, interval[0], interval[1], alpha=0.2, linewidth=0.)
        
    if xlabel is not None:
        ax.set_xlabel(xlabel, fontsize=AXIS_FONT_SIZE)
    if ylabel is not None:
        ax.set_ylabel(ylabel, fontsize=AXIS_FONT_SIZE)
    return ax
```

```python

algorithms = ['ppo', 'trpo', 'dqn', 'dqndueling' ]
names = ['PPO', 'TRPO', 'DDQN', 'Duel-DQN'] 
ls = ['-' , '--' , '-.' , ':']
col = ['r', 'b', 'g', 'orange']
conf = 0.8
n = 5
i = 0
n_running_mean = 500
ax, ax1, ax2, ax3, ax4, ax5 = None, None, None, None, None, None
for alg in algorithms:
    paths = glob.glob("./highway/logs/" + alg + "/progress*.csv")
    paths_sumo = glob.glob("./highway/logs/" + alg + "/progress-sumo*.csv")
    paths = list(set(paths) - set(paths_sumo))
    dfs = [pd.read_csv(p, error_bad_lines=False) for p in paths]
    dfs_sumo = [pd.read_csv(p, error_bad_lines=False) for p in paths_sumo]
    
    if alg in ['dqn', 'dqndueling', 'dqnpr_exp_reply']:
        EpRewMean = [df["mean 100 episode reward"].values.ravel() for df in dfs]
        TimestepsSoFar = [df.steps.values.ravel() for df in dfs]
    else:
        EpRewMean = [df.EpRewMean.values.ravel() for df in dfs]
        TimestepsSoFar = [df.TimestepsSoFar.values.ravel() for df in dfs]
 
    for df in dfs_sumo:
        df['time_steps_cum'] = np.cumsum(df['time_steps']).values.ravel()
    collisions = [running_mean(df['crash'].values.ravel().astype('int'), n_running_mean) for df in dfs_sumo]
    avg_speed = [running_mean(df['speed'].values.ravel(), n_running_mean) for df in dfs_sumo]
    lane_changes = [running_mean((df['lane'].values.ravel() > 0).astype('int'), n_running_mean) for df in dfs_sumo]
    over_limit = [running_mean(df['over_limit'].values.ravel(), n_running_mean) for df in dfs_sumo]
    time_steps = [df['time_steps_cum'].values.ravel() for df in dfs_sumo]
    distance = [running_mean(df['distance'].values.ravel(), n_running_mean) for df in dfs_sumo]
    
    ax = plot_ci(TimestepsSoFar, EpRewMean, conf, n, xlabel='Timesteps', ylabel='Reward', label=names[i], ax=ax, linewidth=2.0, linestyle=ls[i], color=col[i])
    ax1 = plot_ci(time_steps, collisions, conf, n, xlabel='Timesteps', ylabel='Collisions', label=names[i], ax=ax1, linewidth=2.0, linestyle=ls[i], color=col[i])
    ax2 = plot_ci(time_steps, avg_speed, conf, n, xlabel='Timesteps', ylabel='Speed', label=names[i], ax=ax2, linewidth=2.0, linestyle=ls[i], color=col[i])
    ax3 = plot_ci(time_steps, lane_changes, conf, n, xlabel='Timesteps', ylabel='First lane', label=names[i], ax=ax3, linewidth=2.0, linestyle=ls[i], color=col[i])
    ax4 = plot_ci(time_steps, over_limit, conf, n, xlabel='Timesteps', ylabel='Over limit', label=names[i], ax=ax4, linewidth=2.0, linestyle=ls[i], color=col[i])
    ax5 = plot_ci(time_steps, distance, conf, n, xlabel='Timesteps', ylabel='Distance', label=names[i], ax=ax5, linewidth=2.0, linestyle=ls[i], color=col[i])
    
    i += 1
    
ax.legend(loc='lower right', fontsize=LEGEND_FONT_SIZE, ncol=2)
ax.tick_params(axis='x', colors='k', labelsize=TICKS_FONT_SIZE)
ax.tick_params(axis='y', colors='k', labelsize=TICKS_FONT_SIZE)
plt.figure(figs[0].number)
plt.tight_layout()  
plt.savefig('pdfs/' + 'highway_' + 'Reward' + '.pdf')

ax1.legend(loc='upper right', fontsize=LEGEND_FONT_SIZE, ncol=2)
ax1.tick_params(axis='x', colors='k', labelsize=TICKS_FONT_SIZE)
ax1.tick_params(axis='y', colors='k', labelsize=TICKS_FONT_SIZE)
plt.figure(figs[1].number)
plt.tight_layout()  
plt.savefig('pdfs/' + 'highway_' + 'Collisions' + '.pdf')
    
ax2.legend(loc='upper right', fontsize=LEGEND_FONT_SIZE, ncol=2)
ax2.tick_params(axis='x', colors='k', labelsize=TICKS_FONT_SIZE)
ax2.tick_params(axis='y', colors='k', labelsize=TICKS_FONT_SIZE)
plt.figure(figs[2].number)
plt.tight_layout()  
plt.savefig('pdfs/' + 'highway_' + 'Speed' + '.pdf')

    
ax3.legend(loc='upper right', fontsize=LEGEND_FONT_SIZE, ncol=2)
ax3.tick_params(axis='x', colors='k', labelsize=TICKS_FONT_SIZE)
ax3.tick_params(axis='y', colors='k', labelsize=TICKS_FONT_SIZE)
plt.figure(figs[3].number)
plt.tight_layout()  
plt.savefig('pdfs/' + 'highway_' + 'First_lane' + '.pdf')
    
ax4.legend(loc='upper right', fontsize=LEGEND_FONT_SIZE, ncol=2)
ax4.tick_params(axis='x', colors='k', labelsize=TICKS_FONT_SIZE)
ax4.tick_params(axis='y', colors='k', labelsize=TICKS_FONT_SIZE)
plt.figure(figs[4].number)
plt.tight_layout()  
plt.savefig('pdfs/' + 'highway_' + 'Over_limit' + '.pdf')
    
ax5.legend(loc='lower right', fontsize=LEGEND_FONT_SIZE, ncol=2)
ax5.tick_params(axis='x', colors='k', labelsize=TICKS_FONT_SIZE)
ax5.tick_params(axis='y', colors='k', labelsize=TICKS_FONT_SIZE)
plt.figure(figs[5].number)
plt.tight_layout()  
plt.savefig('pdfs/' + 'highway_' + 'Distance' + '.pdf')
    



```

```python
titles = ['reward', 'collisions', 'speed', 'first_lane', 'over_limit', 'distance']

for f, t in zip(figs, titles):
    plt.tight_layout()
    plt.savefig(f, 'pdfs/' + 'highway_' + t + '.pdf')
```
